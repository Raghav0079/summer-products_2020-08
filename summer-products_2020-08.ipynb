```python
# Analyze missing values and document findings

print("=" * 80)
print("MISSING VALUE ANALYSIS")
print("=" * 80)

# Calculate percentage of missing values
missing_percentage = (null_counts / len(df)) * 100
missing_df = pd.DataFrame({
    'Column': null_counts.index,
    'Missing_Count': null_counts.values,
    'Missing_Percentage': missing_percentage.values
}).sort_values('Missing_Percentage', ascending=False)

print("\nüìä Missing Values Summary (sorted by percentage):\n")
print(missing_df[missing_df['Missing_Count'] > 0].to_string(index=False))

print("\n" + "=" * 80)
print("FINDINGS & POSSIBLE REASONS:")
print("=" * 80)

print("\n1Ô∏è‚É£ MERCHANT PROFILE PICTURE (85.4% missing)")
print("   ‚Üí Many merchants may not have uploaded a profile picture")
print("   ‚Üí This is optional information and not critical for product listing")

print("\n2Ô∏è‚É£ HAS_URGENCY_BANNER & URGENCY_TEXT (69.6% missing)")
print("   ‚Üí These fields are only populated when a product has an urgency banner")
print("   ‚Üí Missing values indicate products WITHOUT urgency marketing tactics")
print("   ‚Üí This is intentional missing data (not applicable)")

print("\n3Ô∏è‚É£ RATING BREAKDOWN (2.8% missing)")
print("   ‚Üí rating_five_count, rating_four_count, etc. have 43 missing values")
print("   ‚Üí Likely products with very few or no ratings yet")
print("   ‚Üí Could also be new products without sufficient review data")

print("\n4Ô∏è‚É£ PRODUCT_COLOR (2.7% missing)")
print("   ‚Üí Some products may not have color variations")
print("   ‚Üí Could be items where color is not a relevant attribute")

print("\n5Ô∏è‚É£ ORIGIN_COUNTRY (1.0% missing)")
print("   ‚Üí Minor data entry issue or products without declared origin")

print("\n6Ô∏è‚É£ PRODUCT_VARIATION_SIZE_ID (0.9% missing)")
print("   ‚Üí Some products may not have size variations")
print("   ‚Üí One-size-fits-all products or items where size doesn't apply")

print("\n7Ô∏è‚É£ MERCHANT_NAME & MERCHANT_INFO_SUBTITLE (<0.2% missing)")
print("   ‚Üí Minimal missing data, likely data entry errors")

print("\n" + "=" * 80)
print("RECOMMENDATIONS:")
print("=" * 80)
print("‚úì Rating breakdowns: Can be filled with 0 (no ratings in that category)")
print("‚úì Urgency fields: Can be filled with 0/False and 'No urgency' respectively")
print("‚úì Product color/size: Consider 'Not Specified' or leave as is for analysis")
print("‚úì Merchant profile picture: Can be filled with 0 or 'No Picture'")
print("‚úì Other fields: Investigate individually or use appropriate imputation")
```

```python
# Calculate statistical summary for numerical columns
numerical_stats = df.describe().T

# Add count of non-null values for all columns (including non-numerical)
all_counts = df.count()

# Create a comprehensive statistics dataframe
stats_summary = pd.DataFrame({
    'count': all_counts,
    'mean': df.mean(numeric_only=True),
    'std': df.std(numeric_only=True),
    'min': df.min(numeric_only=True),
    '25%': df.quantile(0.25, numeric_only=True),
    '50%': df.quantile(0.50, numeric_only=True),
    '75%': df.quantile(0.75, numeric_only=True),
    'max': df.max(numeric_only=True)
})

print("=" * 100)
print("STATISTICAL SUMMARY FOR ALL COLUMNS")
print("=" * 100)
print(stats_summary)
```

```python
# Analyze the difference between mean and median for numerical columns

print("=" * 100)
print("MEAN vs MEDIAN ANALYSIS")
print("=" * 100)

# Calculate mean and median for numerical columns
mean_median_df = pd.DataFrame({
    'Mean': df.mean(numeric_only=True),
    'Median': df.median(numeric_only=True)
})

# Calculate the difference and percentage difference
mean_median_df['Difference'] = mean_median_df['Mean'] - mean_median_df['Median']
mean_median_df['Percent_Diff'] = ((mean_median_df['Mean'] - mean_median_df['Median']) / mean_median_df['Median'] * 100).round(2)

# Add skewness indicator
mean_median_df['Skewness_Direction'] = mean_median_df['Difference'].apply(
    lambda x: 'Right-skewed (Positive)' if x > 0.1 else ('Left-skewed (Negative)' if x < -0.1 else 'Symmetric')
)

# Sort by absolute percentage difference
mean_median_df['Abs_Percent_Diff'] = abs(mean_median_df['Percent_Diff'])
mean_median_df_sorted = mean_median_df.sort_values('Abs_Percent_Diff', ascending=False)

print("\nüìä Mean vs Median Comparison (sorted by percentage difference):\n")
print(mean_median_df_sorted[['Mean', 'Median', 'Difference', 'Percent_Diff', 'Skewness_Direction']].to_string())

print("\n" + "=" * 100)
print("KEY FINDINGS & INTERPRETATIONS:")
print("=" * 100)

# Identify highly skewed columns
highly_skewed = mean_median_df_sorted[abs(mean_median_df_sorted['Percent_Diff']) > 50]

print("\nüî¥ HIGHLY SKEWED COLUMNS (>50% difference):\n")
for column in highly_skewed.index:
    mean_val = highly_skewed.loc[column, 'Mean']
    median_val = highly_skewed.loc[column, 'Median']
    pct_diff = highly_skewed.loc[column, 'Percent_Diff']
    
    print(f"\nüìå {column.upper()}")
    print(f"   Mean: {mean_val:.2f} | Median: {median_val:.2f} | Difference: {pct_diff:.2f}%")
    
    # Provide specific interpretations
    if column == 'merchant_rating_count':
        print("   ‚Üí REASON: Few popular merchants have extremely high rating counts (up to 2M+)")
        print("   ‚Üí IMPACT: Most merchants have modest rating counts (~8K), but outliers pull mean up significantly")
        print("   ‚Üí BUSINESS INSIGHT: Market dominated by few established sellers with huge customer bases")
    
    elif column == 'units_sold':
        print("   ‚Üí REASON: Some bestseller products have sold 100K units, while most sell 1K or less")
        print("   ‚Üí IMPACT: Distribution heavily right-skewed with extreme outliers")
        print("   ‚Üí BUSINESS INSIGHT: 'Winner-takes-all' phenomenon - few viral products dominate sales")
    
    elif column == 'rating_count':
        print("   ‚Üí REASON: Popular products accumulate thousands of ratings, but most have few hundred")
        print("   ‚Üí IMPACT: A small number of highly-reviewed products skew the average upward")
        print("   ‚Üí BUSINESS INSIGHT: Product popularity follows power-law distribution")
    
    elif 'rating_five_count' in column or 'rating_four_count' in column or 'rating_three_count' in column or 'rating_two_count' in column or 'rating_one_count' in column:
        print("   ‚Üí REASON: Products with many total ratings naturally have higher individual rating counts")
        print("   ‚Üí IMPACT: Directly correlated with overall rating_count skewness")
        print("   ‚Üí BUSINESS INSIGHT: Rating distribution follows overall review volume pattern")
    
    elif column == 'retail_price':
        print("   ‚Üí REASON: While most products are priced low ($10 median), some luxury items reach $252")
        print("   ‚Üí IMPACT: High-priced outliers significantly inflate the mean")
        print("   ‚Üí BUSINESS INSIGHT: Platform caters mainly to budget-conscious buyers, with few premium products")
    
    elif column == 'countries_shipped_to':
        print("   ‚Üí REASON: Most products ship to ~40 countries, but some reach 140 countries")
        print("   ‚Üí IMPACT: Global shipping strategies vary widely among merchants")
        print("   ‚Üí BUSINESS INSIGHT: International reach is not uniform across sellers")

print("\n" + "=" * 100)
print("üü° MODERATELY SKEWED COLUMNS (10-50% difference):\n")

moderately_skewed = mean_median_df_sorted[
    (abs(mean_median_df_sorted['Percent_Diff']) >= 10) & 
    (abs(mean_median_df_sorted['Percent_Diff']) <= 50)
]

for column in moderately_skewed.index:
    mean_val = moderately_skewed.loc[column, 'Mean']
    median_val = moderately_skewed.loc[column, 'Median']
    pct_diff = moderately_skewed.loc[column, 'Percent_Diff']
    
    print(f"üìå {column}: Mean={mean_val:.2f}, Median={median_val:.2f}, Diff={pct_diff:.2f}%")

print("\n   ‚Üí These columns show moderate skewness, indicating some outliers but not extreme")

print("\n" + "=" * 100)
print("üü¢ SYMMETRIC/NEARLY SYMMETRIC COLUMNS (<10% difference):\n")

symmetric = mean_median_df_sorted[abs(mean_median_df_sorted['Percent_Diff']) < 10]

for column in symmetric.index:
    print(f"üìå {column}: Nearly symmetric distribution")

print("\n   ‚Üí These columns have balanced distributions with minimal outlier influence")

print("\n" + "=" * 100)
print("STATISTICAL RECOMMENDATIONS:")
print("=" * 100)
print("‚úì For highly skewed data (units_sold, rating_count, merchant_rating_count):")
print("  - Use MEDIAN as the central tendency measure for reporting")
print("  - Consider log transformation for modeling and visualization")
print("  - Identify and analyze outliers separately")
print("\n‚úì For moderately skewed data:")
print("  - Both mean and median can be reported for completeness")
print("  - Be aware of outlier influence when making business decisions")
print("\n‚úì For symmetric data:")
print("  - Mean is a reliable measure of central tendency")
print("  - Standard statistical methods apply without major concerns")
```

### üîü Tags Analysis

```python
# Countplot: Top Tags
plt.figure(figsize=(14, 7))

# Extract all tags and count them
all_tags = []
for tags in df['tags'].dropna():
    # Split tags by comma and strip whitespace
    tag_list = [tag.strip() for tag in str(tags).split(',')]
    all_tags.extend(tag_list)

# Count and get top 20 tags
from collections import Counter
tag_counts = Counter(all_tags)
top_tags = dict(tag_counts.most_common(20))

# Create bar plot
plt.barh(list(top_tags.keys()), list(top_tags.values()), color='coral', edgecolor='black', alpha=0.7)
plt.title('Top 20 Product Tags', fontsize=14, fontweight='bold')
plt.xlabel('Count', fontsize=12)
plt.ylabel('Tags', fontsize=12)
plt.gca().invert_yaxis()
plt.grid(axis='x', alpha=0.3)
plt.tight_layout()
plt.show()

print("Top 10 Tags:")
for i, (tag, count) in enumerate(list(top_tags.items())[:10], 1):
    print(f"{i}. {tag}: {count}")
```

### 1Ô∏è‚É£9Ô∏è‚É£ Discount Analysis

```python
# Calculate discount percentage
df['discount_percentage'] = ((df['retail_price'] - df['price']) / df['retail_price'] * 100).clip(0, 100)

# Histogram: Discount Percentage Distribution
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# Discount percentage
axes[0].hist(df['discount_percentage'].dropna(), bins=50, color='crimson', edgecolor='black', alpha=0.7)
axes[0].set_title('Distribution of Discount Percentages', fontsize=14, fontweight='bold')
axes[0].set_xlabel('Discount (%)', fontsize=12)
axes[0].set_ylabel('Frequency', fontsize=12)
axes[0].axvline(df['discount_percentage'].mean(), color='yellow', linestyle='--', linewidth=2, 
               label=f'Mean: {df["discount_percentage"].mean():.1f}%')
axes[0].axvline(df['discount_percentage'].median(), color='lime', linestyle='--', linewidth=2, 
               label=f'Median: {df["discount_percentage"].median():.1f}%')
axes[0].legend()
axes[0].grid(axis='y', alpha=0.3)

# Discount bins
df['discount_bin'] = pd.cut(df['discount_percentage'], 
                            bins=[0, 10, 25, 50, 75, 100], 
                            labels=['0-10%', '10-25%', '25-50%', '50-75%', '75-100%'])

discount_counts = df['discount_bin'].value_counts().sort_index()
axes[1].bar(range(len(discount_counts)), discount_counts.values, 
           color='darkred', edgecolor='black', alpha=0.7)
axes[1].set_xticks(range(len(discount_counts)))
axes[1].set_xticklabels(discount_counts.index, rotation=0)
axes[1].set_title('Products by Discount Range', fontsize=14, fontweight='bold')
axes[1].set_xlabel('Discount Range', fontsize=12)
axes[1].set_ylabel('Number of Products', fontsize=12)
axes[1].grid(axis='y', alpha=0.3)

plt.tight_layout()
plt.show()

print(f"Average Discount: {df['discount_percentage'].mean():.2f}%")
print(f"Median Discount: {df['discount_percentage'].median():.2f}%")
print(f"\nDiscount Range Distribution:")
print(discount_counts)
```
